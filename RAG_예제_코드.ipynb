# Chunking
문서 분할하기 - 고정 크기 청킹(Fixed Size Chunking)<br>
"Hello my name is Jane, and I love technology. This is an example of RAG"<br>
- chunk_size
- chunk_overlap
- chunk_size = 5, chunk_overlap = 1
  - ["Hello my name is Jain", "Jane, and I love technology.", "technology. This is an example", "example of RAG"]

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=256, chunk_overlap=32)
split_documents = text_splitter.split_documents(docs)
```

- Chunking: 큰 문서를 처리하기 쉬운 문단, 문장, 청크 단위로 분할하는 과정 (문장을 잘 나누는 것이 RAG의 성능에 큰 영향을 줌)
- Recursive Chunking: 가장 중요한 구분 기호 > 중요하지 않은 기호 순으로 정렬한 후 분할

```python
print(len(split_documents)) # 121개의 문서가 794개의 Chunk로 분할
```

# Embedding 생성
- 임베딩이란 텍스트를 기계가 이해할 수 있도록 숫자로 이루어진 벡터로 변환하는 과정을 말합니다.

```python
import os
from langchain_openai import OpenAIEmbeddings

OPENAI_API_KEY = ""
os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY
embeddings = OpenAIEmbeddings()
```

```python
# 벡터DB에 임베딩 저장
from langchain_community.vectorstores import FAISS

vectorstore = FAISS.from_documents(documents=split_documents, embedding=embeddings)
```

- Vector DB: ChromaDB, LanceDB, Milvus, Pinecone, ...
- Vector-Capable NoSQL DB: MongoDB, Neo4J, Cassandra, Redis, ...

```python
vectorstore.save_local('./db/faiss')
```

다시 저장된 벡터를 불러서 사용할 때에는, 아래 코드를 입력해주시면 됩니다.<br>
또한 벡터 생성 시에 사용한 것과 동일한 임베딩 모델을 사용해야합니다. <br>  
> vectorstore = FAISS.load_local('./db/faiss', embeddings)

# Retrieval (검색)

```python
faiss_retriever = vectorstore.as_retriever(search_type="similarity", search_kwargs={"k": 3})
```

```python
question = "2024년 국내 GDP 성장률 전망이 어떻게 돼?"
retrieved_chunks = faiss_retriever.invoke(question)
```

```python
print(retrieved_chunks[0].page_content)
```

```python
print(retrieved_chunks[1].page_content)
```

```python
print(retrieved_chunks[2].page_content)
```

```python
# 모든 결과를 하나의 텍스트로 만들기
context = '\n\n'.join(retrieved_chunk.page_content for retrieved_chunk in retrieved_chunks)
print(context)
```

# Generator

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate.from_template("""
  You are a world class expert in 2024 Korea's Economy.
  Use the following chunks of information to answer the question.
  If you don't know the answer, just say that you don't know and do not generate false information.

  Question: {question}
  Context: {context}
  Answer:
""")
```
